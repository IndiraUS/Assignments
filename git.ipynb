### Version Control
        
    Introduction
                
    A project typically consists of many developers
    therefore it is necessary to have a system that can
    track changes.
    
    Version control is the management of changes to
    documents, computer programs, or other files.
    
    Version control tracks every modification to a
    shared project, allowing isolation of issues and
    making prevention and fixes easy and efficient.
                
    For software, version control maintains a central,
    functional codebase that others can copy and make
    additions to.
    
    New code can be written in isolation, leaving the
    original unchanged.
                
    All changes can be reviewed & tested before
    integration into the master
    Changes to one part of the code can be made without
    altering impacting others.
    
    Though many version control systems exist, the popular
    standard for software collaboration in current market is Git.
    
    Following are the types of VCS −
    1.Centralized version control system (CVCS).
    
    Centralized version control system (CVCS) uses a central server to store all files and enables team collaboration.
    But the major drawback of CVCS is its single point of failure, i.e., failure of the central server. Unfortunately, 
    if the central server goes down for an hour, then during that hour, no one can collaborate at all. 
    And even in a worst case, if the disk of the central server gets corrupted and proper backup has not been taken, 
    then you will lose the entire history of the project. Here, distributed version control system (DVCS) comes into picture.
    
    2.Distributed/Decentralized version control system (DVCS).
    
    DVCS clients not only check out the latest snapshot of the directory but they also fully mirror the repository. 
    If the server goes down, then the repository from any client can be copied back to the server to restore it. 
    Every checkout is a full backup of the repository. 
    Git does not rely on the central server and that is why you can perform many operations when you are offline.
    You can commit changes, create branches, view logs, and perform other operations when you are offline.
    
    Git is a distributed version control and source code management system with an emphasis on speed.

    Wiki: https://en.wikipedia.org/wiki/List_of_version-control_software




    


Git is a distributed version control system designed to track changes to a project (code) in software development. It is intended to enforce coordination, collaboration, speed, and efficiency among developers.5. No need of powerful hardware.

### Common Terminologies 

Repositories
                
        A repository is a virtual storage location for a project.
        A git repository is effectively a new directory.
        This directory keeps track of different versions of a project
                
        A project generally contains multiple repositories.
                
        Remote repositories, from which developers obtain new code and changes.
        Local repositories, used by developers to write and test new code.
                
        Both remote and local repositories are functionally the same,
        but usually contain different permissions to protect the
        remote from unnecessary errors.
           
           
           
Branches
                
        A branch is divergent, isolated history of changes 
        used to develop new code in isolation from the 
        master or main body of code.
                
        Every repository has one base, or master branch,
        from which new branchs are made.
        
        Each new branch is created from a specific commit snapshot of the master branch.
        Subsequent changes (called commits) are tracked as part of this new branch.
        Changes do not affect the master branch until a merge is performed.
               
Workflow
                
        Git allows for many different workflows,but most development follows the same structure.
        A project will have one central remote repository
        
        Each developer will create their own local repository
        as a clone of the remote.
                
        The master branch is the current up-to-date version
        of the code.
        Every new code,feature, bug fix, or alteration is done
        within a new branch.
        Upon completion and testing, the feature branch is
        merged into master.
        
        This cycle continues, isolating new development from
        master until it has been thoroughly tested and approved.
                
        
        
        
Merge Conflicts
                
        When the master branch has not changed,
        integrating new code from a different branch is easy.
        
        If both master and the new branch have changed,
        this conflict can cause confusion for git.
        This clash of changes will result in a merge conflict
                
        Git can resolve some merge conflicts automatically,
        if they result from changes in different files.
        When this happens, git will integrate the new changes
        into the current branch.
        
        If the same files or lines have been edited, git will
        need human intervention to figure out what the "correct"
        version is, or to integrate the two together.
                
        Merge conflicts are only visible to the developer
        trying to integrate their new code.
        The rest of the team is unaffected, allowing conflicts to happen with minimal impact

### Git Functionality
 
Cloning
        
        The clone command creates a copy of a given target git repository
        Most commonly, this is used with a remotely hosted repository to give a developer a local copy to work with
        clone needs one argument, the target repository, but can take a second, the destination
                
        If no destination is specified, the copy will exist in the current directory
        user@cpu:~/projectname $ git clone git://repoaddress.git
                
        We can specify the --branch option to only clone a specific branch
        user@cpu:~/projectname $ git clone --branch <branchname> git://repoaddress.git
                
              
Initialization
                
        To create an empty git repository, we use the init command
        user@cpu:~/projectname $ git init
                
        The repository will be created in our current directory,here within the directory "projectname".
        This repository is simply create a new subdirectory, named .git
        
        Since most git commands require an existing repository to
        reference, this is normally the first command you will run.
                
        Directory contents (if they exist) will otherwise be unchanged.
        Once our repository is created, the status command will give
        us a snapshot of the repository's current status. $ git status 
                
Branching
                
        To create a new branch from the current master,
        we can use the branch command 
        
        With no arguments, git branch will simply list all
        branches in our repository and indicate our current branch
        user@cpu:~/projectname $ git branch
                
        If given a new branch name, git branch will create a new branch
        user@cpu:~/projectname $ git branch <branchname>
                
        To switch between branches while working on a codebase,
        we can use the checkout command.
        
        When checking out a different branch, the current working
        directory will be updated to reflect the new branch's history
        
        Subsequent commits will be recorded to the new branch
               
        The checkout command only needs one argument, the branch's name
            user@cpu:~/projectname $ git checkout <branchname>
            checkout can also create a branch before switching to it, with the -b option
            user@cpu:~/projectname $ git checkout -b <new branch>
        
        Delete branch local -```git branch -D <branch_name>```
        If you want to delete branch remotely - ```git push origin --delete <branchname>```
               
               
Staging Area
       
        The add command is used to add changes to the staging area
        add takes one or more arguments, the filenames to be committed
                
            user@cpu:~/projectname $ git add path/to/file.extension
            
            git status should then show us our new file is staged on branch master
            
            user@cpu:~/projectname$ git status                    
            
            ```Your branch is up to date with 'origin/master'.
                Changes to be committed: (use "git reset HEAD ..." to unstage)

                modified:   path/to/file.extension```
                
               
        The commit command takes all changes to the staged files,
        and updates the project history with those changes.
        
        The commit command can be used with a few options:
        -m denotes a message to describe the changes in the commit
                
        All commit messages should be imperative tense(Sit down and eat your lunch.In this sentence,someone is giving a command)
                
        The use of git commit with no options will open the system's
        default text editor, to allow writing of a commit message.
        
        user@cpu:~/projectname $ git commit -m "descriptive message here"
                [master eb33d93] descriptive message here
                1 file changed, 1 insertion(+), 1 deletion(-)
                
        git status should then show us our new commit
                
        user@cpu:~/projectname$ git status
                ```On branch master
                Your branch is ahead of 'origin/master' by 1 commit.```
                
       The stash command saves changes from the current working directory for later use
       These changes are not committed, and must be re-applied later to be useful
                
       This allows a developer to work on other commits without committing current, unrelated changes
       user@cpu:~/projectname $ git stash
       ```Saved working directory and index state WIP on master: eb33d82 descriptive message here```
                
         git status should then show us no staged changes
         
         To stage stashed changes, we can use:
          user@cpu:~/projectname $ git stash pop
                
        
    
Syncing
        
    git push
        
        Once we have committed changes, the push command will send our updated code to the remote repository
        This keeps the remote codebase in sync with local changes, and will ultimately allow local changes to be integrated into the project
                
        user@cpu:~/projectname $ git push <remote name> <branch name> 
        ``` git push origin master / simply git push```
        
        If you are working on changes to a file on a seperate branch, push the contents to the remote repo by performing command 'git push origin <branch_name>'
                
        git status should then show that our branch is up to date with master
            user@cpu:~/projectname$ git status
       
 
        
    git fetch
    
        The fetch command downloads new data (files, commits) from the remote repository
        The fetched data is not merged into the local codebase
                
        Local development is unchanged, the fetched content must be checked out for viewing
        fetch is an easy way to preview new changes before integration into the codebase
                
        The syntax for fetch is simple
        user@cpu:~/projectname $ git fetch <remote name> ```git fetch origin```
                
        Fetch can optionally take a branch name, and only fetch remote commits for that branch
        user@cpu:~/projectname $ git fetch <remote name> <branch name> ``` git fetch origin <branchname>```
                
        Do a git status which will prompt you to do a git pull if merges need to be down between remote and local
     
    git pull
    
        The pull command is actually a combination of fetch followed by merge
        This is primarily used to keep local code in sync with remote changes
                
        With no arguments, pull will simply pull the remote's version of the current branch
        user@cpu:~/projectname $ git pull
                
        We can specify a remote repository if one has not been configured
        user@cpu:~/projectname $ git pull <remote name> ```git pull origin```
        
https://www.freecodecamp.org/news/git-fetch-vs-pull/
    
    git merge
    
        Git's merge command is used to integrate commits from one branch into another
        merge needs one argument, the name of a branch other than the current checked-out branch
        user@cpu:~/projectname $ git merge <other_branch>
                
        The merge command will then find a previous commit that both branches have in common
        From there, git will combine the sequence of commits for each branch into one unified commit history
                
        If successful, these will be added to the current (usually master) branch as a new merge commit
        If unsuccessful, git will require user intervention to resolve the confilict(s)
                
        Conflicts happen when the same data has been changed in both branches
        Git cannot make assumptions about which commit is "correct", so the merge will fail
